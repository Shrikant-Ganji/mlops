```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Step 1: Load the January 2023 Yellow Taxi Trip Data
df = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet")
```

```python
# Step 2: Compute the trip duration in minutes
df["duration"] = (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]).dt.total_seconds() / 60
```

```python
# Step 3: Filter out durations that are outliers
df = df[(df["duration"] >= 1) & (df["duration"] <= 60)]
```

```python
# Step 4: Define the categorical features
categorical = ["PULocationID", "DOLocationID"]
```

```python
# Step 5: Convert to string (DictVectorizer needs string keys)
df[categorical] = df[categorical].astype(str)
```

```python
# Step 6: Convert to dictionary format for one-hot encoding
train_dicts = df[categorical].to_dict(orient="records")
```

```python
# Step 7: One-hot encode
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
# Step 8: Define the target
y_train = df["duration"].values
```

```python
# Step 9: Train the Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Step 10: Predict and calculate RMSE
y_pred = lr.predict(X_train)
rmse = mean_squared_error(y_train, y_pred, squared=False)
print("Q5: Training RMSE:", round(rmse, 2))
```
